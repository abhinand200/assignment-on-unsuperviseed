```python
# Scale the data
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
df_sc=sc.fit_transform(df)
df_sc=pd.DataFrame(df_sc)
```

```python
df_sc.head(10)
```

```python
# Performing PCA(principle component analysis)
from sklearn.decomposition import PCA
pca=PCA(n_components=2)
df_sc=pca.fit_transform(df_sc)
df_pc=pd.DataFrame(df_sc,columns=['pc1','pc2'])
df_pc
```

```python
plt.figure(figsize=(10,8))
sns.scatterplot(x='pc1',y='pc2',data=df_pc)
plt.show()
```

```python
# defining x
x=df_pc
```

**KMeans clustering**

```python
from sklearn.cluster import KMeans
wcss =[]
for i in range (1,11):
  kmeans=KMeans(n_clusters=i,init='k-means++',random_state=42)
  kmeans.fit(x)
  wcss.append(kmeans.inertia_)
```

```python
wcss
```

```python
# check elbow method
plt.plot(range(1,11),wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# the elbow bend show on portion 3
kmeans=KMeans(n_clusters=3,init='k-means++',random_state=42,n_init='auto')
y_pred_km=kmeans.fit_predict(x)
```

```python
y_pred_km
```

```python
from sklearn.metrics import silhouette_score
silhouette_km= silhouette_score(x,y_pred_km)
print(f'silhouette_score kmeans:{silhouette_km}')
```

**AgglomerativeClustering**

```python
from scipy.cluster.hierarchy import linkage,dendrogram
z=linkage(x,method='ward')
```

```python
plt.figure(figsize=(8,4))

plt.axhline(20,color='r',linestyle='--')  #horizondal line

dendrogram(z)
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters=3, metric='euclidean', linkage='ward')
y_hc =hc.fit_predict(x)
```

```python
y_hc
```

```python
from sklearn.metrics import silhouette_score
silhouette_ag = silhouette_score(x,y_hc)
print('AgglomerativeClustering:',silhouette_ag)
```

**DB SCAN**

```python
df=np.array(df)
```

```python
from sklearn.cluster import DBSCAN
dbscan = DBSCAN(eps=3 ,min_samples=2)
dbscan.fit_predict(df)
clusters = dbscan.labels_
clusters
```

```python
from sklearn.metrics import silhouette_score
silhouette_db = silhouette_score(df_sc, clusters)
print("Silhouette Score:", silhouette_db)
```
